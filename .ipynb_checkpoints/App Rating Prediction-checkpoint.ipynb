```python
import numpy as np
import pandas as pd
import re

#Loading the data file
app_ratings_df = pd.read_csv("googleplaystore.csv")

print("Shape of data frame", app_ratings_df.shape)
app_ratings_df.info()
```

```python
#Number of null values for each column
for col in app_ratings_df.columns:
    print(col, "column have",len(app_ratings_df[col]) - app_ratings_df[col].count()," null values")
```

```python
#Drop records(rows) with null values in any column
app_ratings_df.dropna(how='any',inplace=True)

print("Shape of data frame", app_ratings_df.shape)
app_ratings_df.info()
```

```python
#function to replace size mentioned in Mb to Kb
def Convert_size(Size_val):
    if re.match('^\d*[\.\d*]+M$',str(Size_val)):
        converted_val = re.findall('\d+',str(Size_val))
        if(len(converted_val) == 1):
            converted_val = int(converted_val[0])
        elif(len(converted_val) == 2):
            converted_val = float(str(converted_val[0] + '.' + converted_val[1]))
        converted_val = converted_val * 1000
       # converted_val = str(converted_val) + 'k'
        return converted_val
    else:
        return 0


app_ratings_df['Size'] = app_ratings_df['Size'].apply(Convert_size)

app_ratings_df = app_ratings_df[app_ratings_df['Size'] > 0 ] #since app cannot have invalid size or 0

print("The first 5 values of Size column\n",app_ratings_df['Size'][:5])
```

```python
#Converting review field datatype to numeric(int/float)
app_ratings_df['Reviews'] = pd.to_numeric(app_ratings_df['Reviews'])

print("The datatype of Reviews column is",app_ratings_df['Reviews'].dtype)
```

```python
#formating installs fields values
def Convert_installs(Install_val):
    if (isinstance(Install_val, str) and re.match('[\,\+]?',Install_val)):
        Install_val = Install_val.replace(",", "")        #removing "," from the string
        Install_val = Install_val.replace("+","")         #removing "+" from the string
        return int(Install_val)                           #coverting string to int
    else:
        return Install_val
    
app_ratings_df['Installs'] = app_ratings_df['Installs'].apply(Convert_installs)
print("The first 5 values of intalls field is\n",app_ratings_df['Installs'][:5])
```

```python
#formating Price column values

#removing the '$' from the fields values of Price column
app_ratings_df['Price'] = app_ratings_df['Price'].str.replace("$","")

#converting the data type of Price column to numeric
app_ratings_df['Price'] = pd.to_numeric(app_ratings_df['Price'])

print("The unique values of Price column\n", app_ratings_df['Price'].unique())
```

```python
#identifying the rows which has the ratings values greater than 5 and removing them.

#filter for finding ratings less than 5
ratings_below_5 = app_ratings_df['Rating'] < 5

#applying filter on the data frame
app_ratings_df = app_ratings_df[ratings_below_5]

#fetching unique values to validate
app_ratings_df['Rating'].unique()
```

```python
#identifying the rows which have reviews greater than installs

print("shap of df before applying filter",app_ratings_df.shape)

#filter for the findings row which has review less than or equal to installs
filter_reviews_installs = app_ratings_df['Reviews'] <= app_ratings_df['Installs']

app_ratings_df = app_ratings_df[filter_reviews_installs]

print("shap of df after applying filter",app_ratings_df.shape)
```

```python
#identifying the rows which are type free and cost >0

print("shap of df before applying filter",app_ratings_df.shape)

#filter for Type free
filter_type_free = app_ratings_df['Type'] == "Free"

#filter for cost equal to zero
filter_cost_zero = app_ratings_df['Price'] > 0.0

app_ratings_df = app_ratings_df[~(filter_cost_zero & filter_type_free)]

print("shap of df before applying filter",app_ratings_df.shape)
```
